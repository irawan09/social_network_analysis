```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
```

```python
graph = nx.Graph()

graph.add_edge('Steven', 'Indra', weight=25)
graph.add_edge('Indra', 'Adam', weight=17)
graph.add_edge('Jonathan', 'Cassey', weight=15)
graph.add_edge('Indra', 'Gabriel', weight=16)
graph.add_edge('Jonah', 'Jeremy', weight=24)
graph.add_edge('Casey', 'Adam', weight=30)
graph.add_edge('Adam', 'Jonathan', weight=27)
graph.add_edge('Jeremy', 'Jonah', weight=25)
graph.add_edge('Gabriel', 'Chaos', weight=19)
graph.add_edge('Adam', 'Gabriel', weight=18)
graph.add_edge('Adam','Jeremy', weight=10)
graph.add_edge('Indra','Eeve', weight=8)
graph.add_edge('Chaos', 'Jonah', weight=10)

print(nx.info(graph))
```

```python
plt.figure(figsize=(20,20))
nx.spring_layout(graph)
nx.draw_networkx(graph);
```

```python
# Measure the degree of node connections
print("Number of degre (Adam) : "+ str(nx.degree(graph, 'Adam')))
```

```python
# Create the shortest path to certain node
print("Shortest path from Gabriel to Jonathan are (based on weight) : ")


shortest_path = nx.shortest_path(graph, 'Gabriel', 'Jonathan')
for x in range(len(shortest_path)):
    print(x, str(shortest_path[x]))
```

```python
# Using Breath-first Search algorithm for finding the shortest path starting from Jonah
bfs = nx.bfs_tree(graph, 'Jonah')
plt.figure(figsize=(20,20))
nx.spring_layout(bfs)
nx.draw_networkx(bfs)
```
