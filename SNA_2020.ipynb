```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
```

```python
graph_1 = nx.DiGraph()

graph_1.add_edge('Claire', 'Fred')
graph_1.add_edge('Fred', 'Claire')
graph_1.add_edge('Claire', 'Susan')
graph_1.add_edge('Susan', 'Claire')
graph_1.add_edge('Fred', 'Sarah')
graph_1.add_edge('Sarah', 'Fred')
graph_1.add_edge('Susan', 'Sarah')
graph_1.add_edge('Sarah', 'Susan')
graph_1.add_edge('Susan', 'Tom')
graph_1.add_edge('Tom', 'Susan')
graph_1.add_edge('Tom', 'David')
graph_1.add_edge('David', 'Tom')
graph_1.add_edge('David', 'Steven')
graph_1.add_edge('Steven', 'David')
graph_1.add_edge('Susan', 'Steven')
graph_1.add_edge('Tom','Steven')
graph_1.add_edge('Steven','Tom')
graph_1.add_edge('Susan','David')
graph_1.add_edge('Fred', 'Susan')
graph_1.add_edge('Claire', 'Sarah')
graph_1.add_edge('Sarah', 'Claudia')
graph_1.add_edge('Claudia', 'Sarah')
graph_1.add_edge('Steven', 'Claudia')
graph_1.add_edge('Claudia', 'Ben')
graph_1.add_edge('Ben', 'Jennifer')
graph_1.add_edge('Jennifer', 'Ben')

print(nx.info(graph_1))
```

```python
plt.figure(figsize=(10,10))
nx.spring_layout(graph_1)
nx.draw_networkx(graph_1);
```

```python
reciprocity = nx.overall_reciprocity(graph_1)

print ("Reciprocity of this graph 1 : ", reciprocity)
```

```python
degree_dict = dict(graph_2.degree(graph_2.nodes()))
nx.set_node_attributes(graph_2, degree_dict, 'degree')

print("Romul_10 data : ", graph_2.nodes['Romul_10'])
```
